```python
# A project that helps automize searching for books in a library
```

```python
import csv 
import webbrowser

def book_name(): #taking in user input for bookname
    global bookname 
    bookname = input("Enter the books name: ") 
    
def check(): #checking if book is in the excel sheet
    data=open('example.csv',encoding='utf-8') 
    csv_data=csv.reader(data) 
    data_lines=list(csv_data) 
    num= (len(data_lines)) 
    i=num-1
    
    while num>=0 and i>=0 and num>=0:
        if str(data_lines[i][0]).lower() == str(bookname).lower():
            print (bookname+" is in the library, ask your librarian for status updates.") 
            break 
        elif (str(data_lines[i][0]).lower() != str(bookname).lower()): 
            i=i-1 
            num=num-1 
        
    if (str(data_lines[i][0]).lower() != str(bookname).lower()):#if book not found, giving link to buy
        try:
            from googlesearch import search 
        except ImportError: 
            print("No module named 'google' found") 
            
        print("The book is not in the library. You could buy it if you like: ") 
                             
        for j in search("buy book "+bookname, tld="co.in", num=1, stop=1, pause=1):
                             webbrowser.open(j) # opening link on new tab
```

```python
book_name()
check()
```
